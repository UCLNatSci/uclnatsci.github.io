```python
import matplotlib.image as mpi
img = mpi.imread('atlantic-basin.png')

plt.imshow(img)

# latitude/longitude co-ordinates of edges of map
left = -90
right = -17.06
bottom = 0
top = 45

# dimensions of image in pixels
width = 964
height = 600

#lat/long of New York

lat_NY = 40.7
long_NY = -74.0

x_NY = (long_NY + 90)*964/(right-left)
y_NY = (height*(1 - (lat_NY - bottom)/(top-bottom)))

print("NY x pos:", x_NY)
print("NY y pos:", y_NY)

plt.scatter(x_NY, y_NY)
plt.text(x_NY, y_NY, "New York", color="white")


```

```python
import csv

x = []
y = []

# Open the data file
with open("irma.csv") as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        latitude = float(row[2])
        longitude = float(row[3])
        
        x.append((longitude + 90)*964/(right-left))
        y.append((height*(1 - (latitude - bottom)/(top-bottom)))
        
print("x-coords:", x)
print("y-coords:", y)
```

```python
plt.imshow(img)
plt.scatter(x, y, color='purple')
```

```python
from matplotlib import animation

ims = []
for i in range(10):
    im = plt.imshow(img, animated=True)
    plt.scatter(i*90, 300)
    ims.append([im])
    
    
fig = plt.figure()
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.image as mpi
img = mpi.imread('atlantic-basin.png')

fig = plt.figure()
ax = plt.axes()
patch = plt.Circle((0, 0), 20, fc='y')


def init():
    patch.center = (20, 20)
    ax.add_patch(patch)
    return patch,

def animate(i):
    patch.center = (x[i], y[i])
    return patch,

anim = animation.FuncAnimation(fig, animate, 
                               init_func=init, 
                               frames=len(x), 
                               interval=20,
                               blit=True)

plt.imshow(img,zorder=0)
anim.save('hurricane_irma.mp4', writer = 'ffmpeg', fps=30)
plt.show()
```
